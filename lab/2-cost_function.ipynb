## Computing Cost

The equation for cost with one variable is:
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$

where
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{2}$$


- $f_{w,b}(x^{(i)})$ is our prediction for example $i$ using parameters $w,b$.
- $(f_{w,b}(x^{(i)}) -y^{(i)})^2$ is the squared difference between the target value and the prediction.
- These differences are summed over all the $m$ examples and divided by `2m` to produce the cost, $J(w,b)$.


```python
def compute_cost(x, y, w, b):
    """
    Computes the cost function for linear regression.

    Args:
      x (ndarray (m,)): Data, m examples
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters

    Returns
        total_cost (float): The cost of using w,b as the parameters for linear regression
               to fit the data points in x and y
    """
    # number of training examples
    m = x.shape[0]

    cost_sum = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum = cost_sum + cost
    total_cost = (1 / (2 * m)) * cost_sum

    return total_cost
```

## OOP in Matplotlib

Matplotlib’s `pyplot` module acts as a **global state manager** that tracks the active **Figure** and **Axes**, forwarding drawing commands to the underlying object-oriented components.

- The **Figure** (`matplotlib.figure.Figure`) serves as the top-level canvas container, holding all subplots and controlling the global layout settings for the entire visualization.

- An **Axes** (`matplotlib.axes._axes.Axes`) instance represents an individual plot or subplot with its own coordinate system, where the actual plotting logic (such as lines, markers, and annotations) is executed.

## Useful Drawing APIs

### Create Figure and Axes

`fig, ax = plt.subplots(1, 2, constrained_layout=True, figsize=(8, 4))`: Create a figure `fig` containing one row and two subplots `ax[]` with specified figure size and automatic compact layout adjustment.

### Axe Label & Text Configuration

`ax[0].set_title("Title: Left Axes")`: Set the top title text for the left subplot.

`ax[0].set_ylabel("Y Label")`: Add a descriptive label to the vertical Y-axis of the left subplot.

`ax[0].set_xlabel("X Label")`: Add a descriptive label to the horizontal X-axis of the left subplot.

`ax[0].text(0.4, 0.05, "TEXT", transform=a[0].transAxes, color="red")`: Place fixed red text at the relative position (40% width, 5% height) inside the left subplot canvas.

### Plot Elements Drawing

`ax[0].scatter([1,2,3,4], [300,400,700,550], marker='x', c='r', label="Label: Scatter")`: Draw red cross-shaped discrete scatter points on the left subplot and define its legend name.

`ax[0].plot([1,2,3,4], [300,400,700,550], label="Label: Plot")`: Draw a continuous polyline connecting given coordinate points on the left subplot and set its legend label.

`ax[1].hlines(y=2, xmin=0, xmax=5, color="gray", linestyle="--", label="Label: Hlines")`: Draw a gray dashed horizontal reference line at Y=2 across the specified X range with a legend label.

`ax[1].vlines(x=2.5, ymin=0, ymax=6, color="blue", ls=":", label="Label: VLines")`: Draw a blue dotted vertical reference line at X=2.5 spanning the specified Y range and assign a legend name.

`ax[1].annotate("CrossPoint",xy=(2.5,2),xytext=(4,5),arrowprops=dict(arrowstyle="->", color="red"))`: Attach text annotation with a red arrow pointing to the target data coordinate on the right

### Legend Rendering

`ax[0].legend()`: Show a legend box on the left subplot to present labels of scatter and line series.
