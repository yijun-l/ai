```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

`np.zeros()` creates a NumPy array filled entirely with zeros according to the specified shape.

`compute_model_output(x, w, b)` compute the value of $f_{w,b}(x^{(i)})$

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples
      w,b (scalar)    : model parameters
    Returns
      y (ndarray (m,)): target values
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

`plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')` draws the model prediction line with blue color and a legend tag.

`plt.legend()` shows the legend box to match lines/points with their labels.

```python
tmp_f_wb = compute_model_output(x_train, w, b,)

plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

Through manual computation, the fitted coefficients are $w = 200$ and $b = 100$
